# Styling with Cycler

```python
from cycler import cycler
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
```

```python
print(yy.shape)
```

```python
yy[:, 3]
```

```python
# cycler 사용의 기본 형태
default_cycler = (cycler(color=['r','g','b','y']) + cycler(linestyle=['-', '--', ':', '-.']))
```

```python
plt.rc('lines', linewidth=4)
# runtime current에서 prop_cycle로도 불러 낼 수 있다.
plt.rc('axes', prop_cycle=default_cycler) 
```

```python
custom_cycler = (cycler(color=['c', 'm', 'y', 'k']) +
                 cycler(lw=[1, 2, 3, 4]))
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2)
ax0.plot(yy)
ax0.set_title('Set default color cycle to rgby')
#set_prop_cycle이 cycler를 쓸수 있게 해준다.
ax1.set_prop_cycle(custom_cycler) 
ax1.plot(yy)
ax1.set_title('Set axes color cycle to cmyk')

fig.subplots_adjust(hspace=0.5)
plt.show()
```

# Cycling through multiple properties

```python
from cycler import cycler
cc = (cycler(color=list('rgb')) + cycler(linestyle=['-', '--', '-.']))
for d in cc :
    print(d)
    
# cycler 더하기의 경우 단순히 r + - / g + -- / b + -. 의 형태를 만들어준다.
```

```python
mul_cc = (cycler(color=list('rgb')) * cycler(linestyle=['-', '--', '-.']))

for d in mul_cc : 
    print(d)

# cycler 곱하기의 경우 r * -, --, -. 의 식으로 앞의 것과의 곱을 만든다. 
```
