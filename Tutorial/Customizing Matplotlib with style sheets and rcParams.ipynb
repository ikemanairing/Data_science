# Customizing Matplotlib with style sheets and rcParams

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
from cycler import cycler #from cycler이다.
```

# Runtime rc setting

```python
# rcParams는 Default를 건드린다.

mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '--'

data = np.random.randn(50)
plt.plot(data)
plt.show()
```

```python
# change default plot color with cycler!

mpl.rcParams['axes.prop_cycle'] = cycler(color=['r', 'g', 'b', 'y']) # cycler는 key, value에서 value를 차례로 바꿔준다
plt.plot(data) # first color is red
```

```python
# Shortcut to rc

mpl.rc('lines', linewidth=4, linestyle='-.')
plt.plot(data)
```

# Temporary rc setting

```python
# Temporary rc setting 

with mpl.rc_context({'lines.linewidth': 2, 'lines.linestyle': ':'}) :
    plt.plot(data) #with에 함께 넣는다. 
```

```python
# 데코레이터로 새로 정의 해도 된다.

@mpl.rc_context({'lines.linewidth': 3, 'lines.linestyle': '--'})
def plotting_function() :
    plt.plot(data)
    plt.show()
```

```python
plotting_function()
```
