```python
fig = plt.figure() # 바탕이 되는 figure 마련
fig.subplots_adjust(top=0.8) # 위쪽을 늘린다. 
ax1 = fig.add_subplot(211) # ax1은 2,1,1의 서브플롯이다.
ax1.set_ylabel('volts') # y라벨은 volts
ax1.set_title('a sine wave') # 타이틀은 a sine wave

t = np.arange(0.0, 1.0, 0.01) # 0~1사이의 0.01로 잘게 나누어진 구간
s = np.sin(2*np.pi*t) # 2 * pi * 0.01에서 2 pi 까지의 사인 값
line, = ax1.plot(t, s, color='blue', lw=2) # x값은 t, y값이 s인 파란색 두께 2의 선

np.random.seed(19680801) # 랜덤 구간 저장

ax2 = fig.add_axes([0.126, 0.1, 0.778, 0.33]) # 이건 ticks나 label이 아니야. ax의 left bottom, width, height 값이야
n, bins, patches = ax2.hist(np.random.randn(1000), 50, facecolor='yellow', edgecolor='yellow')
ax2.set_xlabel('time (s)') # 

plt.show()
```

```python
import matplotlib.lines as lines

fig = plt.figure() #figure를 만들고

l1 = lines.Line2D([0, 1], [0, 1], transform = fig.transFigure, figure=fig) #
l2 = lines.Line2D([0, 1], [1, 0], transform = fig.transFigure, figure=fig)
fig.lines.extend([l1, l2])

plt.show()
```
